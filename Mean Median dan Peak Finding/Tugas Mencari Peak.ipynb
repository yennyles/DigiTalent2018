```python
import numpy as np
import statistics as stats
import random
import matplotlib.pyplot as plt
import scipy
import scipy.signal as signal

X =np.array([3,7,9,6,5,8,4,7,5])
print ("Data Asli \n", X)

index,_=signal.find_peaks(X, distance=2)

print("\n Peak : ",index)

plt.plot(X)
plt.plot(index,X[index],"x")
plt.show()
```
